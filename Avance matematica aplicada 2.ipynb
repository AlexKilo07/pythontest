```python
from scipy.optimize import fsolve

def M(x):
    y=32
    return 0.5*x+2-y

x_inicial=0

x_solucion=fsolve(M,x_inicial)

print(f"El numero de usuario de usuarios activos que provoca un us de memoria de 32 GB es: {x_solucion}")
```

```python
from scipy.optimize import fsolve
import numpy as np

def A(t): 
    return 0.36*t**2 

def B(t):
    return 0.16*t**2+t+10 

def diferencia(t):
    return A(t)-B(t)

solucion=fsolve(diferencia,5)[0]

print(f"Dias de trabajo: {solucion:.2f}")
```

```python
from scipy.optimize import fsolve

def I(x):
    return 10*x**2+50*x 

def C(x):
    return 5*x**2+80*x+100

def ecuacion(x):
    y=385
    return C(x)-y

x_inicial=5

solucion=fsolve(ecuacion,x_inicial)[0]

solucion2=I(3)

print(f"La cantidad de suscriptores es: {solucion} mil")
print(f"El ingreso es de {solucion2} mil")
```

```python
from scipy.optimize import fsolve

def Sactual(x):
    return 600*1.05**x

total=Sactual(4)

print(f"El sueldo actual luego de vender 4 casas es: {total:.2f} mil")

def solucion(x):
    y=930.797
    return 600*1.05**x-y

x_inicial=1

solucion_x=fsolve(solucion,x_inicial)[0]

print(f"La cantidad de casas vendidas es: {solucion_x:.2f}")
```

```python
import matplotlib.pyplot as plt
import numpy as np

def f(x):
    return 500000+6000*x

def g(x):
    return 600000 +4000*x

x=np.arange(0, 100)

plt.plot(x,f(x),label="f(x)")
plt.plot(x,g(x),label="g(x)")

plt.title("Grafico de posibles formulas")
plt.xlabel("Numero de ventas por mes")
plt.ylabel("Sueldo a pagar en pesos")

plt.grid(True)
plt.legend()
plt.show()

```
